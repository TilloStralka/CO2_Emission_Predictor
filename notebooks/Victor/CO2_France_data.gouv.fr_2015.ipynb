Additional-Info on "cod_cbr" - fuel type: <br>
EE: petrol/electric (rechargeable) <br>
EH: petrol/electric (non-rechargeable) <br>
EL: electric <br>
ES: unleaded petrol 95 <br>
ES/GN: dual-fuel petrol/Natural Gas Vehicle – NGV (petrol consumption data) <br>
ES/GP: dual-fuel petrol/Liquefied Petroleum Gas – LPG (petrol consumption data) <br>
FE: E85 vehicle (or superethanol-E85) <br>
GH: diesel/electric (non-rechargeable) <br>
GL: diesel/electric (rechargeable) <br>
GN: single-fuel Natural Gas Vehicle – NGV <br>
GN/ES: dual-fuel petrol/Natural Gas Vehicle – NGV (NGV consumption data) <br>
GO: diesel <br>
GP/ES: dual-fuel petrol/Liquefied Petroleum Gas – LPG (LPG consumption data) <br>
H2: hydrogen <br>


```python
#load packages, libs
import chardet
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# set the option to display all columns
pd.set_option('display.max_columns', None)
```

```python
# store file-path of dataset in own var
file_path = "/content/drive/MyDrive/CO2/data_fic_etiq_edition_40-mars-2015.csv"

# Read a small portion of the file to detect the encoding
with open(file_path, "rb") as f:
    result = chardet.detect(f.read(100000))  # Read the first 100,000 bytes

# use result to set encoding when reading/loading the dataset
print(result["encoding"])
```

```python
# load/read datset into df
df = pd.read_csv(file_path, sep=';', encoding = "ISO-8859-1", decimal=".")
```
