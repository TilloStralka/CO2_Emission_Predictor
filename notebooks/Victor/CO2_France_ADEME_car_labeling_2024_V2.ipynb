# First Look into ADEME - car - labelling data for -> CO2-emissions of passenger cars (France only)

Purpose: <br>
The siteCar Labelingprovides an analytical view of the new market.

ADEME also provides source data, reworked from various sources, to enable the emergence of new applications that will encourage the acquisition of vehicles with the least impact or additional analyses to inform decision-makers and purchasers.
<br>
<br>

Dataset includes all type (not every singel!) of registered passengers cars in France for 2024 (started March 2024). <br>
(Dataset is ~1MB big: 3604 rows, 52 columns)<br>
(after duplicates removal: 2309 rows, 52 columns)

Data- source
- link: https://www.data.gouv.fr/fr/datasets/ademe-car-labelling/

```python
# store file-path of dataset in own var#
# update to your path...
file_path = "/content/drive/MyDrive/CO2/data_France_ADEME-car_labeling_2024.csv"

# Read a small portion of the file to detect the encoding
with open(file_path, "rb") as f:
    result = chardet.detect(f.read(100000))  # Read the first 100,000 bytes

# use result to set encoding when reading/loading the dataset -> (UTF-8-SIG)
print(result["encoding"])
```

```python
# show unique types of energy type
print(df_renamed["Energy"].unique())

# dictionary to map French classes to English translations
translation_dict_energy = {
    "ESSENCE": "GASOLINE",
    "ELECTRIC": "ELECTRIC",
    "GAZOLE": "DIESEL",
    "ELEC+ESSENC HR": "ELECTRIC+GASOLINE HYBRID",
    "GAZ+ELEC HNR": "GAS+ELECTRIC HYBRID",
    "ESS+ELEC HNR": "GASOLINE+ELECTRIC HYBRID",
    "SUPERETHANOL": "SUPERETHANOL",
    "ESS+G.P.L.": "GASOLINE+LPG",
    "ELEC+GAZOLE HR": "ELECTRIC+DIESEL HYBRID"
}

# replace French names with English in the "energy" column
df_renamed["Energy"] = df_renamed["Energy"].replace(translation_dict_energy)
print()

# show unique types of energy type
print(df_renamed["Energy"].unique())
```

```python
# show unique types of energybodywork type
print(df_renamed["Bodywork"].unique())

# dictionary to map French bodywork classes to English translations
translation_dict_bodywork = {
    "COMBISPACE": "COMPACT MPV",
    "TS TERRAINS/CHEMINS": "SUV/ALL-TERRAIN",
    "MONOSPACE COMPACT": "COMPACT MINIVAN",
    "BERLINE": "SEDAN",
    "BREAK": "STATION WAGON",
    "COUPE": "COUPE",
    "CABRIOLET": "CONVERTIBLE",
    "MINIBUS": "MINIBUS",
    "MINISPACE": "SMALL MPV",
    "MONOSPACE": "MINIVAN"
}

# replace French names with English in the 'Bodywork' column
df_renamed["Bodywork"] = df_renamed["Bodywork"].replace(translation_dict_bodywork)
print()

# show unique types of energy type
print(df_renamed["Bodywork"].unique())
```

```python
# show unique types of Range type
print(df_renamed["Range"].unique())

# dictionary to map French range classes to English translations
translation_dict_range = {
    "INFERIEURE": "LOWER",
    "MOYENNE INFERIEURE": "LOWER MIDDLE",
    "MOYENNE SUPERIEURE": "UPPER MIDDLE",
    "LUXE": "LUXURY",
    "SUPERIEURE": "UPPER",
    "ECONOMIQUE": "ECONOMY"
}

# replace French names with English in the 'Range' column
df_renamed["Range"] = df_renamed["Range"].replace(translation_dict_range)
print()

# show unique types of energy type
print(df_renamed["Range"].unique())
```

```python
# show unique Gearbox_Type
print(df_renamed["Gearbox_Type"].unique())

# dictionary to map French Gearbox types to English translations
translation_dict_gearbox = {
    "AUTOMATIQUE": "AUTOMATIC",
    "VARIATION ELECTRIQUE": "ELECTRIC VARIABLE",
    "MECANIQUE": "MANUAL",
    "VARIATION CONTINUE": "CONTINUOUS VARIABLE"
}

# replace French names with English in the 'Gearbox_type' column
df_renamed["Gearbox_Type"] = df_renamed["Gearbox_Type"].replace(translation_dict_gearbox)
print()

# show unique types of Gearbox_type after translation
print(df_renamed["Gearbox_Type"].unique())
```

```python
# show unique types of Bonus_Malus
print(df_renamed["Bonus_Malus"].unique())

# Dictionary to map French Bonus/Malus types to English translations
translation_dict_bonus_malus = {
    "Malus": "Penalty",
    "Bonus 6000": "Bonus 6000",
    "Neutre 0": "Neutral 0",
    "Bonus 4000": "Bonus 4000",
    "Bonus 1000": "Bonus 1000",
    "Bonus 2000": "Bonus 2000",
    "AUTRES": "Others"
}

# Replace French names with English in the 'Bonus_Malus' column
df_renamed["Bonus_Malus"] = df_renamed["Bonus_Malus"].replace(translation_dict_bonus_malus)
print()

# Show unique types of Bonus_Malus after translation
print(df_renamed["Bonus_Malus"].unique())
```

```python
# now some plots/graphs

# create pd.series/ calc: value counts for brands
brand_count = df_renamed["Brand"].value_counts()

# plot car brands
plt.figure(figsize=(10, 6))
sns.barplot(x = brand_count.index, y = brand_count.values, order = brand_count.index)
plt.title("Number of car Brands registered 2024")
plt.xticks(rotation=90)
plt.xlabel("Brand")
plt.ylabel("Count")
plt.show;
```

```python

# get the counts of each energy type and sort by count
energy_counts = df_renamed["Energy"].value_counts().index

# sns countplot of energy type
plt.figure(figsize=(6, 4))
sns.countplot(x='Energy', data=df_renamed, order = energy_counts)
plt.title("Energy Type Distribution")
plt.xlabel("Energy Type")
plt.ylabel("Count")
plt.xticks(rotation=90)
plt.show();


```

```python
# Pivot the data to create a table of counts for each Energy type per Brand
stacked_data = df_renamed.pivot_table(index="Brand", columns="Energy", aggfunc='size', fill_value=0)

# Sort brands by total counts
stacked_data = stacked_data.loc[stacked_data.sum(axis=1).sort_values(ascending=False).index]

# Plot the stacked bar chart
stacked_data.plot(kind='bar', stacked=True, figsize=(12, 6), width = 0.9)
plt.title("Stacked Bar Chart of Energy Type Distribution by Brand")
plt.xlabel("Brand")
plt.ylabel("Count")
plt.xticks(rotation=90)
plt.legend(title="Energy Type")
plt.show()
```
